## API Call

```python
%%time

# Code adapted from: 
# https://github.com/whatismybrowser/api-v2-sample-code/blob/master/sample-code/python-3.6/user_agent_parse.py

# Create empty arrays for each data field
hardware_type = []
operating_system_name = []
software_sub_type = []
operating_system_version = []
simple_sub_description_string = []
software_type = []
hardware_sub_sub_type = []
extra_info = []
software_version_full = []
software_name_code = []
layout_engine_name = []
capabilities = []
operating_system_flavour_code = []
detected_addons = []
operating_system_flavour = []
operating_system_frameworks = []
hardware_sub_type = []
software_name = []
software_version = []
simple_operating_platform_string = []
layout_engine_version = []
extra_info_dict = []
operating_platform_vendor_name = []
operating_system = []
operating_system_version_full = []
operating_platform_code = []
operating_system_name_code = []
operating_platform = []
simple_software_string = []
user_agent = []
software = []

# Request will be sent to
api_url = "https://api.whatismybrowser.com/api/v2/user_agent_parse"

# -- Set up HTTP Headers
headers = {
    'X-API-KEY': 'YourKeyHere',
}

count = 0
while (count < len(browsers)):
    post_data = {'user_agent': str(browsers['user_agents'][count]),}
    # -- Make the request
    result = requests.post(api_url, data=json.dumps(post_data), headers=headers)

    # -- Try to decode the api response as json
    result_json = {}
    try:
        result_json = result.json()
    except Exception as e:
        print(result.text())
        print("Couldn't decode the response as JSON:", e)
        exit()

    # -- Check that the server responded with a "200/Success" code
    if result.status_code != 200:
        print("ERROR: not a 200 result. instead got: %s." % result.status_code)
        print(json.dumps(result_json, indent=2))
        exit()

    # -- Check the API request was successful
    if result_json.get('result', {}).get('code') != "success":
        print("The API did not return a 'success' response. It said: result code: %s, message_code: %s, message: %s" % (result_json.get('result', {}).get('code'), result_json.get('result', {}).get('message_code'), result_json.get('result', {}).get('message')))
        #print(json.dumps(result_json, indent=2))
        exit() 

    # Now you have "result_json" and can store, display or process any part of the response.

    # -- print the entire json dump for reference
    #print(json.dumps(result_json, indent=2))

    # -- Copy the data to some variables for easier use
    parse = result_json.get('parse')
    version_check = result_json.get('version_check')

    # Now you can do whatever you need to do with the parse result
    # Store results in individual arrays, if no result, store an empty value
    try:
        hardware_type.append(parse.get('hardware_type'))
    except:
        hardware_type.append('')
    try:
        operating_system_name.append(parse.get('operating_system_name'))
    except:
        operating_system_name.append('')
    try:
        software_sub_type.append(parse.get('software_sub_type'))
    except:
        software_sub_type.append('')
    try:
        operating_system_version.append(parse.get('operating_system_version'))
    except:
        operating_system_version.append('')
    try:
        simple_sub_description_string.append(parse.get('simple_sub_description_string'))
    except:
        simple_sub_description_string.append('')
    try:
        software_type.append(parse.get('software_type.append'))
    except:
        software_type.append('')
    try:
        hardware_sub_sub_type.append(parse.get('hardware_sub_sub_type'))
    except:
        hardware_sub_sub_type.append('')
    try:
        extra_info.append(parse.get('extra_info'))
    except:
        extra_info.append('')
    try:
        software_version_full.append(parse.get('software_version_full'))
    except:
        software_version_full.append('')
    try:
        software_name_code.append(parse.get('software_name_code'))
    except:
        software_name_code.append('')
    try:
        layout_engine_name.append(parse.get('layout_engine_name'))
    except:
        layout_engine_name.append('')
        
    try:   
        capabilities.append(parse.get('capabilities'))
    except:
        capabilities.append('')
    try:
        operating_system_flavour_code.append(parse.get('operating_system_flavour_code'))
    except:
        operating_system_flavour_code.append('')
    try:
        detected_addons.append(parse.get('detected_addons'))
    except:
        detected_addons.append('')
    try:
        operating_system_flavour.append(parse.get('operating_system_flavour'))
    except:
        operating_system_flavour.append('')
    try:
        operating_system_frameworks.append(parse.get('operating_system_frameworks'))
    except:
        operating_system_frameworks.append('')
    try:
        hardware_sub_type.append(parse.get('hardware_sub_type'))
    except:
        hardware_sub_type.append('')
    try:
        software_name.append(parse.get('software_name'))
    except:
        software_name.append('')
    try:
        software_version.append(parse.get('software_version'))
    except:
        software_version.append('')
    try:
        simple_operating_platform_string.append(parse.get('simple_operating_platform_string'))
    except:
        simple_operating_platform_string.append('')
    try:
        layout_engine_version.append(parse.get('layout_engine_version'))
    except:
        layout_engine_version.append('')
    try:
        extra_info_dict.append(parse.get('extra_info_dict'))
    except:
        extra_info_dict.append('')
    try:
        operating_platform_vendor_name.append(parse.get('operating_platform_vendor_name'))
    except:
        operating_platform_vendor_name.append('')
    try:
        operating_system.append(parse.get('operating_system'))
    except:
        operating_system.append('')
    try:
        operating_system_version_full.append(parse.get('operating_system_version_full'))
    except:
        operating_system_version_full.append('')
    try:
        operating_platform_code.append(parse.get('operating_platform_code'))
    except:
        operating_platform_code.append('')
    try:
        operating_system_name_code.append(parse.get('operating_system_name_code'))
    except:
        operating_system_name_code.append('')
    try:
        operating_platform.append(parse.get('operating_platform'))
    except:
        operating_platform.append('')
    try:
        simple_software_string.append(parse.get('simple_software_string'))
    except:
        simple_software_string.append('')
    try:
        user_agent.append(parse.get('user_agent'))
    except:
        user_agent.append('')
    try:
        software.append(parse.get('software'))
    except:
        software.append('')
       
    count = count + 1

# Refer to:
# https://developers.whatismybrowser.com/api/docs/v2/integration-guide/#user-agent-parse-field-definitions
# for more fields you can use
```

## Python Library Call

```python
# Code adapted from:
# https://github.com/selwin/python-user-agents

# Create empty arrays for each data field
browser = []
browser_family = []
browser_version = []
browser_version_string = []
browser_is_mobile = []
device_is_tablet = []
device_is_touch_capable = []
device_is_pc = []
device_is_bot = []
browser_string = []
device_os = []
device_os_family = []
device_os_version = []
device_os_version_string = []
device = []
device_family = []
device_brand = []
device_model = []

d = 0
while (d < len(browsers)):
    ua_string = str(browsers['user_agents'][d])
    user_agent = parse(ua_string)
    try:
        browser.append(user_agent.browser)    
    except:
        browser.append('')
    try:
        browser_family.append(user_agent.browser.family)
    except:
        browser_family.append('')
    try:
        browser_version.append(user_agent.browser.version)
    except:
        browser_version.append('')
    try:
        browser_version_string.append(user_agent.browser.version_string)
    except:
        browser_version_string.append('')
    try:
        browser_is_mobile.append(user_agent.is_mobile)
    except:
        browser_is_mobile.append('')
    try:
        device_is_tablet.append(user_agent.is_tablet)
    except:
        device_is_tablet.append('')
    try:
        device_is_touch_capable.append(user_agent.is_touch_capable)
    except:
        device_is_touch_capable.append('')
    try:
        device_is_pc.append(user_agent.is_pc)
    except:
        device_is_pc.append('')
    try:
        device_is_bot.append(user_agent.is_bot)
    except:
        device_is_bot.append('')
    try:
        browser_string.append(str(user_agent))
    except:
        browser_string.append('')
    try:
        device_os.append(user_agent.os)  # returns OperatingSystem(family=u'iOS', version=(5, 1), version_string='5.1')
    except:
        device_os.append('')
    try:
        device_os_family.append(user_agent.os.family)  # returns 'iOS'
    except:
        device_os_family.append('')
    try:
        device_os_version.append(user_agent.os.version)  # returns (5, 1)
    except:
        device_os_version.append('')
    try:
        device_os_version_string.append(user_agent.os.version_string)  # returns '5.1'
    except:
        device_os_version_string.append('')
    try:
        device.append(user_agent.device)  # returns Device(family=u'iPhone', brand=u'Apple', model=u'iPhone')
    except:
        device.append('')
    try:
        device_family.append(user_agent.device.family)  # returns 'iPhone'
    except:
        device_family.append('')
    try:
        device_brand.append(user_agent.device.brand) # returns 'Apple'
    except:
        device_brand.append('')
    try:
        device_model.append(user_agent.device.model)
    except:
        device_model.append('')
        
    d += 1
```

```python
# Fill new dataframe with results from user_agent library

browsers_library['browser'] = browser
browsers_library['browser_family'] = browser_family
browsers_library['browser_version'] = browser_version
browsers_library['browser_version_string'] = browser_version_string
browsers_library['browser_is_mobile'] = browser_is_mobile
browsers_library['device_is_tablet'] = device_is_tablet
browsers_library['device_is_touch_capable'] = device_is_touch_capable
browsers_library['device_is_pc'] = device_is_pc
browsers_library['device_is_bot'] = device_is_bot
browsers_library['browser_string'] = browser_string
browsers_library['device_os'] = device_os
browsers_library['device_os_family'] = device_os_family
browsers_library['device_os_version'] = device_os_version
browsers_library['device_os_version_string'] = device_os_version_string
browsers_library['device'] = device
browsers_library['device_family'] = device_family
browsers_library['device_brand'] = device_brand
browsers_library['device_model'] = device_model
```
